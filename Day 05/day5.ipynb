## Part 1

```python
# file = open("example.txt")
file = open("test.txt")
correct = 143
res = 0

lines = file.readlines()
file.close()

rules = []
updates = []

for l in lines:
    if "|" in l:
        rules.append([int(x) for x in l.strip().split('|')])
    if "," in l:
        updates.append([int(x) for x in l.strip().split(',')])


def is_valid_update(update, rules):
    for i in range(0, len(update)):
        for j in range(i + 1, len(update)): 
            page_i = update[i]
            page_j = update[j]

            for rule in rules:
                if rule[0] == page_i and rule[1] == page_j:
                    # valid: page_i before page_j
                    break
                if rule[0] == page_j and rule[1] == page_i:
                    # invalid: page_j before page_i
                    return False
                
    return True


for update in updates:
    if is_valid_update(update, rules):
        res += update[int(len(update)/2)]
        # pprint(update)

if res == correct:
    print(f"Correct: {res}")
else:
    print(f"Not correct, res {res}, correct {correct}")
```

```python
# file = open("example.txt")
file = open("test.txt")
correct = 123
res = 0
val = 0

lines = file.readlines()
file.close()

rules = []
updates = []

for l in lines:
    if "|" in l:
        rules.append([int(x) for x in l.strip().split('|')])
    if "," in l:
        updates.append([int(x) for x in l.strip().split(',')])


def is_valid_update(update, rules):
    for i in range(0, len(update)):
        for j in range(i + 1, len(update)): 
            page_i = update[i]
            page_j = update[j]

            for rule in rules:
                if rule[0] == page_i and rule[1] == page_j:
                    # valid: page_i before page_j
                    break
                if rule[0] == page_j and rule[1] == page_i:
                    # invalid: page_j before page_i
                    return False
                
    return True

def correct_order(update, rules):
    sorted_update = update.copy()

    ordered = False
    while not ordered:
        ordered = True
        for i in range(len(sorted_update)):
            for j in range(i + 1, len(sorted_update)):
                page_i = sorted_update[i]
                page_j = sorted_update[j]
                
                for rule in rules:
                    if rule[0] == page_i and rule[1] == page_j:
                        continue
                    if rule[0] == page_j and rule[1] == page_i:
                        # Swap pages
                        sorted_update[i], sorted_update[j] = sorted_update[j], sorted_update[i]
                        ordered = False
                        break

    return sorted_update
    

for update in updates:
    if not is_valid_update(update, rules):
        new_update = correct_order(update, rules)
        res += new_update[int(len(update)/2)]
        # pprint(update)

if res == correct:
    print(f"Correct: {res}")
else:
    print(f"Not correct, res {res}, correct {correct}")
```
