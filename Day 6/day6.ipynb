```python
import numpy as np
import re
from pprint import pprint
from enum import Enum
from tqdm import tqdm

class direction(Enum):
    UP = 0
    RIGHT = 1
    DOWN = 2
    LEFT = 3
```

## Part 1

```python
file = open("example.txt")
file = open("test.txt")
correct = 41
res = 0

lines = file.readlines()
file.close()

map_arr = []

for l in lines:
    map_arr.append(list(l.rstrip()))

# pprint(map_arr)

pos_x = -1
pos_y = -1

for x in range(0, len(map_arr)):
    for y in range(0, len(map_arr[0])):
        if map_arr[x][y] == '^':
            pos_x = x
            pos_y = y
            map_arr[x][y] == '.'

def right_rotate(dir):
    return direction((dir.value + 1) % 4)

obstacles = True
dir = direction.UP
memo = set()

while(obstacles):
    if dir == direction.UP:
        for i in range(pos_x-1, -1, -1):
            if map_arr[i][pos_y] == '#':
                pos_x = i + 1
                dir = right_rotate(dir) 
                break
            memo.add(str([i, pos_y]))
            # map_arr[i][pos_y] = "X"
        else:
            obstacles = False
    elif dir == direction.RIGHT:
        for i in range(pos_y+1, len(map_arr[0])):
            if map_arr[pos_x][i] == '#':
                pos_y = i - 1
                dir = right_rotate(dir) 
                break
            memo.add(str([pos_x, i]))
            # map_arr[pos_x][i] = "X"
        else:
            obstacles = False

    elif dir == direction.DOWN:
        for i in range(pos_x+1, len(map_arr)):
            if map_arr[i][pos_y] == '#':
                pos_x = i - 1
                dir = right_rotate(dir) 
                break
            memo.add(str([i, pos_y]))
            # map_arr[i][pos_y] = "X"
        else:
            obstacles = False


    elif dir == direction.LEFT:
        for i in range(pos_y-1, -1, -1):
            if map_arr[pos_x][i] == '#':
                pos_y = i + 1
                dir = right_rotate(dir) 
                break
            memo.add(str([pos_x, i]))
            # map_arr[pos_x][i] = "X"
        else:
            obstacles = False


# pprint(map_arr)
res = len(memo)

if res == correct:
    print(f"Correct: {res}")
else:
    print(f"Not correct, res {res}, correct {correct}")
```

## Part 2

```python
# file = open("example.txt")
file = open("test.txt")
correct = 6
res = 0

lines = file.readlines()
file.close()

map_arr = []

for l in lines:
    map_arr.append(list(l.rstrip()))

# pprint(map_arr)

start_pos_x = -1
start_pos_y = -1

for x in range(0, len(map_arr)):
    for y in range(0, len(map_arr[0])):
        if map_arr[x][y] == '^':
            start_pos_x = x
            start_pos_y = y
            map_arr[x][y] == '.'

def right_rotate(dir):
    return direction((dir.value + 1) % 4)


# map_arr[6][3] = '#'
def check_loop():
    pos_x = start_pos_x
    pos_y = start_pos_y
    obstacles = True
    dir = direction.UP
    memo = set()
    vis = set()
    loop = False
    while(obstacles):
        if dir == direction.UP:
            for i in range(pos_x-1, -1, -1):
                if map_arr[i][pos_y] == '#':
                    pos_x = i + 1
                    dir = right_rotate(dir) 
                    break
                if str([i, pos_y, dir]) in memo:
                    loop = True
                    break
                memo.add(str([i, pos_y, dir]))
                vis.add(str([i, pos_y]))
            else:
                obstacles = False
            if loop:
                break
        elif dir == direction.RIGHT:
            for i in range(pos_y+1, len(map_arr[0])):
                if map_arr[pos_x][i] == '#':
                    pos_y = i - 1
                    dir = right_rotate(dir) 
                    break
                if str([pos_x, i, dir]) in memo:
                    loop = True
                    break
                memo.add(str([pos_x, i, dir]))
                vis.add(str([pos_x, i]))
            else:
                obstacles = False
            if loop:
                break

        elif dir == direction.DOWN:
            for i in range(pos_x+1, len(map_arr)):
                if map_arr[i][pos_y] == '#':
                    pos_x = i - 1
                    dir = right_rotate(dir) 
                    break
                if str([i, pos_y, dir]) in memo:
                    loop = True
                    break
                memo.add(str([i, pos_y, dir]))
                vis.add(str([i, pos_y]))
            else:
                obstacles = False
            if loop:
                break

        elif dir == direction.LEFT:
            for i in range(pos_y-1, -1, -1):
                if map_arr[pos_x][i] == '#':
                    pos_y = i + 1
                    dir = right_rotate(dir) 
                    break
                if str([pos_x, i, dir]) in memo:
                    loop = True
                    break
                memo.add(str([pos_x, i, dir]))
                vis.add(str([pos_x, i]))
            else:
                obstacles = False
            if loop:
                break
    return (loop, vis)

_, new_obs = check_loop()

for n_obs in tqdm(new_obs):
    n_obs = eval(n_obs)
    if n_obs[0] == start_pos_x and n_obs[1] == start_pos_y:
        continue
    map_arr[n_obs[0]][n_obs[1]] = '#'
    loop, _ = check_loop()
    if loop:
        res += 1
    map_arr[n_obs[0]][n_obs[1]] = '.'

# pprint(new_obs)

if res == correct:
    print(f"Correct: {res}")
else:
    print(f"Not correct, res {res}, correct {correct}")
```
