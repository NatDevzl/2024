## Part 1

```python
file = open("example.txt")
correct = 1930
# file = open("example2.txt")
# correct = 140
file = open("test.txt")

res = 0

lines = file.readlines()
file.close()

map_arr = []

directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

for l in lines:
    map_arr.append(list(l.rstrip()))

plant = ""


def fences(x, y, visited):
    plant = map_arr[x][y]
    visited.add((x, y))
    perimeter = 0
    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if nx < 0 or nx >= len(map_arr) or ny < 0 or ny >= len(map_arr[0]) or map_arr[nx][ny] != plant:
            perimeter += 1
            continue
        if (nx, ny) in visited:
            continue
        perimeter += fences(nx, ny, visited)
    return perimeter


total_visited = set()
for x in range(len(map_arr)):
    for y in range(len(map_arr[0])):
        if (x, y) in total_visited:
            continue
        visited = set()
        peri = fences(x, y, visited)
        res += peri * len(visited)
        total_visited.update(visited)


if res == correct:
    print(f"Correct: {res}")

else:
    print(f"Not correct, res {res}, correct {correct}")
```
